```python
import torch
```

## Tensor attributes

```python
f32_tensor = torch.tensor([1,2,3,4,5], 
                          dtype=torch.float32,
                          device=None,
                          requires_grad=False
                         )
```

```python
f32_tensor
```

```python
f16_tensor = f32_tensor.type(torch.float16)
f16_tensor
```

```python
x = f32_tensor * f16_tensor
x, x.dtype
```

```python
x[:, None]
```

```python
y = f32_tensor.type(torch.int32) * f16_tensor.type(torch.int32)
y, y.dtype
```

## Manipulating Tensors (tensor operations)
**Tensor operations** :
* Addition
* Substraction
* Multiplication (element wise)
* Division
* Matrix multiplication

```python
A = torch.tensor([1,2,3])
A
```

```python
A + 10
```

```python
A * 10
```

```python
A/10
```

```python
A = torch.tensor(
    [
        [1,2,3], 
        [4,5,6]
    ]
)
B = torch.tensor(
    [
        [7,8,9],
        [11,12,13]
    ]
)
```

```python
A * B
```

```python
A @ B
```

```python
A @ B.T
```

```python
B @ A.T
```

```python
B.T @ A
```

```python
torch.add(A, 20)
```

```python
torch.mul(B, 2)
```

```python
torch.matmul(A, B.T)
```
